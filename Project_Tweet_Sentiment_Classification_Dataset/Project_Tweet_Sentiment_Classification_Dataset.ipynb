```python
path = "/content/tweet_sentiment.csv"
df = pd.read_csv(path)
df
```

```python
df.info()
```

```python
def transfrom_sentiment_to_number(text):
  if text == "positive":
    return 2
  elif text == "negative":
    return 1
  else:
    return 0
```

```python
df['n_sentiment'] = df['sentiment'].apply(transfrom_sentiment_to_number)
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['tweet'], df['n_sentiment'], test_size = 0.2, random_state = 42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
tf_idf = TfidfVectorizer(stop_words='english')
X_train_tf = tf_idf.fit_transform(X_train.tolist())
X_test_tf = tf_idf.transform(X_test.tolist())
```

```python
X_train_tf.shape, X_test_tf.shape
```

```python
naive_bayes_classifier = MultinomialNB()
naive_bayes_classifier.fit(X_train_tf, y_train)
```

```python
predict = naive_bayes_classifier.predict(X_test_tf)
```

```python
target_names = ['neutral', 'negative', 'positive']
print(metrics.classification_report(y_test, predict,target_names=target_names))
```
